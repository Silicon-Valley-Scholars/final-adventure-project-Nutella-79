```python
import time

def intro():
  print("Welcome - you are unlucky enough to wake up to see a zombie apocalypse!")

intro()

response_selected = False
game = ""
level = ""

while not response_selected:
  game_input = input("Pick easy mode or hard mode: ").lower()

  if game_input == "easy mode":
    game = "easy mode"
    level_chosen = False
    while not level_chosen:
      level_input = input("Pick a level (1-3): ")
      if level_input.isdigit() and 1 <= int(level_input) <= 3:
        level = int(level_input)
        if level == 3:
          print("At least you chose the hardest easy level!")
        else:
          print("You can do better than choose one of the easiest levels!")
        level_chosen = True
        response_selected = True
      else:
        print("Invalid input, please enter a value 1-3.")
  elif game_input == "hard mode":
    game = "hard mode"
    response_selected = True
    print("Good job for not being a scaredy cat, now brace yourself!")
  else:
    print("Invalid input, please enter 'easy mode' or 'hard mode'.")

print("                   --- Game Start ---")

time.sleep(2)

weapons = ["sword", "gun", "axe", "knife", "dagger", "bow and arrow"]

print("let's decide a weapon!")
print("Here are your options:")
print("*note* when typing your response - remember this is case sensitive")
for each in weapons:
  print("-", each)

chosen_weapon = ""
weapon_chosen = False
while not weapon_chosen:
  chosen_weapon_input = input("Choose your weapon: ")
  if chosen_weapon_input in weapons:
    chosen_weapon = chosen_weapon_input
    print("You have chosen your weapon. There's no going back. Good luck.")
    weapon_chosen = True
  else:
    print("That's not a valid weapon. Please pick one from the list.")

time.sleep(2)
print("                   --- Next Step ---")


names = ["Bob", "Bobby", "Bobber", "Bobbit", "Bobbee", "Billy Bob Joe"]

print("Let's decide a name!")
print("Here are your options:")
print("*note* when typing your response - remember this is case sensitive")
for each in names:
  print("-", each)

player_name = ""
names_chosen = False
while not names_chosen:
  chosen_names_input = input("Choose your name: ").title()

  if chosen_names_input in names:
    player_name = chosen_names_input
    print("Welcome. Your adventure begins now.")
    names_chosen = True
  else:
    print("That's not a valid name. Please pick one from the list.")
print("You are all set up!")

def calculate_survival_chance(weapon, difficulty_str):
    base_chance = 50.0

    difficulty_lower = difficulty_str.lower()
    if "hard mode" in difficulty_lower:
        base_chance -= 20
    elif "easy mode level 1" in difficulty_lower:
        base_chance += 15
    elif "easy mode level 2" in difficulty_lower:
        base_chance += 10
    elif "easy mode level 3" in difficulty_lower:
        base_chance += 5

    weapon_lower = weapon.lower()
    if "knife" in weapon_lower:
        base_chance -= 15
    elif "dagger" in weapon_lower:
        base_chance += 3
    elif "axe" in weapon_lower:
        base_chance += 6
    elif "bow and arrow" in weapon_lower:
        base_chance += 9
    elif "sword" in weapon_lower:
        base_chance += 12
    elif "gun" in weapon_lower:
        base_chance += 20

    survival_chance = max(0, min(100, base_chance))
    return survival_chance

time.sleep(2)
print("                   --- Information ---")

if game == "easy mode":
    full_difficulty_string = f"easy mode level {level}"
else:
    full_difficulty_string = "hard mode"

calculated_survival_chance = calculate_survival_chance(chosen_weapon, full_difficulty_string)

player_info = {
    "name": player_name,
    "weapon": chosen_weapon,
    "survival_chance": f"{calculated_survival_chance:.2f}%"
}

print("Player Information:")
print("Name:", player_info["name"])
print("Weapon:", player_info["weapon"])
print("Survival Chance:", player_info["survival_chance"])

time.sleep(3)

print("                                           --- SIKE ---")

print("You really thought I was good enough to make an entire game? Think again! This is the end of your adventure. Buh Bye!!!")


```
